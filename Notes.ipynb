# HISTORY OF WEB

-> Basically World Wide Web (WWW) was developed by Tim Berners Lee in 1989 with a motive that scientists/researchers across the globe can share information with each other. Tim Berners-Lee created HTML in late 1991 but didn't release it officially.He published it later in 1995 as HTML 2.0.

-> World Wide Web is like a collection of interlinked documents. Every document is having hyperlinks which redirect it to some other document and hence this graph like structure is formed.

-> Also, with the help of a web browser, one can view all the documents from anywhere around the world.

-> This was the first website ever made: http://info.cern.ch/hypertext/WWW/TheProject.html which was made using Hypertext Markup Language (HTML). It looked like this:



Then initially after making all these websites in HTML, people wanted to add some style in it, then came CASCADING STYLE SHEETS. Using HTML+CSS, we can make the websites look like the modern world websites, though we cannot make them interactive.

Then after that, to make them interactive, we need to add Javascript to it.

HTML is generally used to design the basic structure of the webpages, which will be then improved using other technologies like CSS and JavaScript. CSS controls the looks, feel, layout and formatting, whereas JavaScript is used to control different elements' behaviour on the page.

# CLIENT SERVER ARCHITECTURE

So basically there are two people involved ::

1. Client Side : Here, basically you have a web browser, on which you type the URL of the page you want to see. That page is present on some system(which is called the server). So the client sends a REQUEST to the server, requesting the server to provide him with the required data.

Similarly, basically more than one client requests the server for the data.

2. Server Side : Here, this always-on system is having information about what to provide to the clients. It basically maintains different or same record for the client (depending upon the website).

![Client Server Architecture](https://github.com/piyushsharma220699/Web-Development/blob/main/Images/clientserverarchitecture.jpg "Client Server Architecture")

If the content that is served to the people is same for everybody, then the website is called STATIC WEBSITE.
However, if the content served is different, it is known as DYNAMIC WEBSITE.

So basically the features of client and server side are:

1. CLIENT SIDE :: This is the front end side and is only concerned about the presentation of the data it gets from the server. It just needs the HTML, CSS & JS file and it will present it.

2. SERVER SIDE :: This is the backend side which is concerned about the data. It is responsible for providing you with the right data.
